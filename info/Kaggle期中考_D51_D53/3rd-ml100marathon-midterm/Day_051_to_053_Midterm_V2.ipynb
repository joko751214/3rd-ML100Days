```python
import pandas as pd
import numpy as np
from IPython.display import display

data_path = "C:/Users/infiark/Desktop/review/GitHub/3rd-ML100Days/info/期中考_D51_D53/3rd-ml100marathon-midterm/"
df_train = pd.read_csv(data_path + "train_data.csv")
df_test = pd.read_csv(data_path + "test_features.csv")
target = pd.read_csv(data_path + "sample_submission.csv")

train_Y = df_train['poi']
ids = df_test['name']
df_train = df_train.drop(['name','poi'], axis=1)
df_test = df_test.drop(['name'], axis=1)
df = pd.concat([df_train, df_test])
df.head()
```

```python
# 檢查 DataFrame 空缺值的狀態
def na_check(df_data):
    data_na = (df_data.isnull().sum() / len(df_data)) * 100
    data_na = data_na.drop(data_na[data_na == 0].index).sort_values(ascending=False)
    missing_data = pd.DataFrame({'missing_ratio': data_na})
    display(missing_data.head(20))
    return list(missing_data[missing_data['missing_ratio'] < 50].index)

# 缺失比小於50的欄位
mlt50 = na_check(df)
```

```python
# 特徵工程
# df = df[mlt50]
df['bonus'] = df['bonus'].map(lambda x: 1 if x > df['bonus'].mean() else 0)
df['salary'] = df['salary'].map(lambda x: 1 if x > df['salary'].mean() else 0)
df['exercised_stock_options'] = df['exercised_stock_options'].map(lambda x: 1 if x > df['exercised_stock_options'].mean() else 0)
df['restricted_stock'] = df['restricted_stock'].map(lambda x: 1 if x > df['restricted_stock'].mean() else 0)
df = df.drop(['email_address', 'total_payments', 'total_stock_value'], axis=1)
df = df.fillna(0)
na_check(df)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV, train_test_split

df_mm = MinMaxScaler().fit_transform(df)

train_num = train_Y.shape[0]
train_X = df_mm[:train_num]
test_X = df_mm[train_num:]

x_train, x_test, y_train, y_test = train_test_split(train_X, train_Y, test_size=0.25, random_state=4)
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error
# 使用RandomSearch尋找最佳參數
tol = np.arange(0.0001, 1, 0.0001)
penalty = ['l1', 'l2']
fit_intercept = [True, False]
C = np.arange(0.01, 1, 0.01)
param_dict = dict(tol=tol, penalty=penalty, fit_intercept=fit_intercept, C=C)

random_search = RandomizedSearchCV(LogisticRegression(), param_dict, n_jobs=-1, verbose=1)

# 開始搜尋最佳參數
acc = 0
while acc < 0.8:
    search_result = random_search.fit(x_train, y_train)
    lr = LogisticRegression(
        tol=search_result.best_params_['tol'], 
        penalty=search_result.best_params_['penalty'],
        fit_intercept=search_result.best_params_['fit_intercept'],
        C=search_result.best_params_['C'])
    
    lr.fit(x_train, y_train)
    y_pred = lr.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(acc)
```

```python
lr.fit(train_X, train_Y)
pred_Y = lr.predict_proba(test_X)[:, 1]
print(pred_Y)
sub = pd.DataFrame({"name": ids, "poi": pred_Y})
sub.to_csv(data_path + "Day_051_to_053_Midterm_V2.csv", index=False)

target = target['poi'].values

mse = mean_squared_error(target, pred_Y)
print("mse:", mse)
```
